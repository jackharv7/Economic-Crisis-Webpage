```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
aapl = data.DataReader("^IXIC", 
                       start='1970-1-1', 
                       end='1985-12-31', 
                       data_source='yahoo')['Adj Close']
normalize = (aapl / aapl.iloc[0] * 100)

nasdaq = pd.DataFrame(normalize)
nasdaq.columns = ["Close"]
nasdaq.to_csv('nasdaq.csv') 
```

```python
unemployment = pd.read_excel("SeriesReport-20200328111051_74441e.xlsx")
unemployment = unemployment.set_index("Year")
#normalize = (unemployment / unemployment.iloc[0] * 100)
mean = unemployment.mean(axis = 1)
mean = list(mean)

unemployment["Mean"] = mean
#unemployment
unemployment["Mean"].plot()
unemployment.to_csv('unemployment.csv') 
```
